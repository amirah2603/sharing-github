```python
# classify sub category type using number

def find_common_words(keywords):
    count1 = 0
    count2 = 0
    count3 = 0
    try:
        for index,row in table_rule.iterrows():
#     print (row['Billing'])
#     print (row['Assurance'])
#     print (row['Fulfillment'])
            if (row['Billing'] != ''):
                if (row['Billing'] in keywords):
                    count1 = count1 + 1
#         print("bill")
            if (row['Assurance'] != ''):
                if (row['Assurance'] in keywords):
#         print("assurance")
                    count2 = count2 + 1
            if (row['Fulfillment'] != ''):
                if (row['Fulfillment'] in keywords):
#         print("fulfillment")
                    count3 = count3 + 1

#         count4 = 0
#         count5 = 0
        
#         match1 = keywords == ''
#         match2 = re.search(sub_case2, keywords) 
#         match3 = re.search(sub_case3, keywords)
#         match4 = re.search(sub_case4, keywords)
        
# #         match1 = keywords == ''
# #         match2 = re.findall(sub_case2,keywords)
# #         match3 = re.findall(sub_case3,keywords)
# #         match4 = re.findall(sub_case4, keywords)
       
#         if match1:
# #             return 1
#             count1 += 1
#         if match2:
# #             return 2
#             count2 += 1
#         if match3:
# #             return 3
#             count3 += 1
#         if match4:
# #             return 4
#             count4 += 1
# #         else:
# #             return 5
# #             count5 = 0
       
        if (count1 > count2):
            if (count1 > count3):
                return 2
            else:
                return 6
        elif (count2 > count3):
            if (count2 > count1):
                return 3
            else:
                return 5
        elif (count3 > count1):
            if (count3 > count2):
                return 4
            else:
                return 7
        elif (count2 != 0):
            if (count2 == count1):
                if (count2 == count3):
                    return 8
        else:
            return np.nan
                    
        
    except TypeError:
        return None
    
    
df["Category Type"] = df["input"].apply(find_common_words)
df
```

```python
# classify new sub category

new_sub = []

for row in df['Category Type']:
    if row == 1:
        new_sub.append('Unknown')
    elif row == 2:
        new_sub.append('Bills')
    elif row == 3:
        new_sub.append('Assurance')
    elif row == 4:
        new_sub.append('Fulfillment')
    elif row == 5:
        new_sub.append('Bills & Assurance')
    elif row == 6:
        new_sub.append('Bills & Fulfillment')
    elif row == 7:
        new_sub.append('Assurance & Fulfillment')
    elif row == 8:
        new_sub.append('All')
    else:
        new_sub.append('Unknown')

        
df['Category_New'] = new_sub
df
```
